```python
def get_col_with_zero_std(df: pd.DataFrame):
    
    drop_col = []
    num_col = [col for col in df.columns if df[col].dtype != "O"]
    for col in num_col:
        if df[col].std() == 0:
            drop_col.append(col)

    return drop_col

def get_redundant_col(df: pd.DataFrame, thresh = 0.7):
    col_missing_ratio = df.isna().sum().div(df.shape[0])
    cols_to_drop = list(col_missing_ratio[col_missing_ratio>thresh].index)

    return cols_to_drop
```

```python
drop_cols_1 = get_col_with_zero_std(df = wafers)
drop_cols_2 = get_redundant_col(df = wafers)
```

```python
drop_cols = drop_cols_1 + drop_cols_2
```
